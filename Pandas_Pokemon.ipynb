```python
import pandas as pd
import numpy as np


df  = pd.read_csv("poke.csv")
df.head(10)
```

```python
df.columns = df.columns.str.upper().str.replace('_',' ') #change column names
df.head(10)
```

```python
df = df.set_index('NAME') # change and set the index to the name attribute
df.head(10)
```

```python
df=df.drop(['#'],axis=1) #drop the columns with axis=1;axis=0 is for rows
df.columns
```

```python
df['TYPE 2'].fillna(df['TYPE 1'], inplace=True)  # fill Nan values in Type2 with Type1
df.head(10)
```
